#CREATE DATASETS to organize data

```python
import torch

class SafetyDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

                #  HARD CHECK (prevents silent mismatch bugs)
        assert len(self.encodings["input_ids"]) == len(self.labels), "Mismatch!"

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

    def __len__(self):
        return len(self.labels)

train_dataset_xlmr = SafetyDataset(train_encodings_xlmr, train_labels)
val_dataset_xlmr   = SafetyDataset(val_encodings_xlmr, val_labels)

train_dataset_mdeberta = SafetyDataset(train_encodings_mdeberta, train_labels)
val_dataset_mdeberta   = SafetyDataset(val_encodings_mdeberta, val_labels)

print("✅ Datasets ready")
```

```python
import torch
# import torch.nn.functional as F
from transformers import Trainer

class WeightedTrainer(Trainer):
    def __init__(self, class_weights, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights

    def compute_loss(self, model, inputs, return_outputs=False):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        # 👇 THIS IS THE EXACT LINE YOU ASKED ABOUT
        loss_fct = torch.nn.CrossEntropyLoss(
            weight=self.class_weights.to(logits.device),
            label_smoothing=0.1   # ← ADD IT HERE
        )

        loss = loss_fct(logits, labels)
        return (loss, outputs) if return_outputs else loss
```

```python
import numpy as np
from transformers import TrainingArguments, Trainer
from sklearn.metrics import f1_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    return {"f1": f1_score(labels, preds, average="macro")}
    
training_args = TrainingArguments(
    output_dir="./results",

    eval_strategy="steps",
    eval_steps=300,
    save_strategy="steps",   
    save_steps=300,          # 👈 match eval_steps
    save_total_limit=1,
    
    learning_rate=3e-5, #better+heavier obv
        # learning_rate=2e-5, 
    lr_scheduler_type="cosine",
    warmup_steps=500,
    max_grad_norm=1.0,

    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=4,   # even faster
    
    num_train_epochs=1,   # keep fast
    
    load_best_model_at_end=True,
    metric_for_best_model="f1",

    weight_decay=0.01,

    logging_steps=50,

    bf16=True,
    fp16=False, #(only if A100 / modern GPU — not always your case)

    dataloader_num_workers=2,
    
    report_to="none"
)

print("✅ Training setup ready")

# only logs during .train()
def train_model(model, train_dataset, val_dataset, tokenizer, name):
    print(f"\n🚀 Training {name}")

    trainer = WeightedTrainer(
        class_weights=class_weights,# ONLY here if your custom class supports it
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        tokenizer=tokenizer,# processing_class=tokenizer,
        compute_metrics=compute_metrics
    ) 

    trainer.train()
    result = trainer.evaluate()

    print(f"🏁 {name} RESULTS:", result)

    return trainer, result
```
